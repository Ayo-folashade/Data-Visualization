```python
# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
# Scatterplot of sepal length versus sepal width, colored by species
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatterplot of Sepal Length vs Sepal Width')
plt.show()
```

```python
# Barplot of average petal length by species
sns.barplot(x='species', y='petal_length', data=iris, errorbar=None)
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.title('Barplot of Average Petal Length by Species')
plt.show()
```

```python
# Pie chart of relative frequencies of species
species_counts = iris['species'].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart of Relative Frequencies of Species')
plt.show()
```

```python
# Subplots of boxplots of sepal length and petal length by species
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
sns.boxplot(x='species', y='sepal_length', data=iris, ax=axes[0])
sns.boxplot(x='species', y='petal_length', data=iris, ax=axes[1])
axes[0].set_xlabel('Species')
axes[0].set_ylabel('Sepal Length (cm)')
axes[0].set_title('Boxplot of Sepal Length by Species')
axes[1].set_xlabel('Species')
axes[1].set_ylabel('Petal Length (cm)')
axes[1].set_title('Boxplot of Petal Length by Species')
plt.tight_layout()
plt.show()
```
